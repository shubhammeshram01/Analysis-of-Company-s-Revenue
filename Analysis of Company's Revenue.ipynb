### Columns like Units sold,Manufacturing Price,Sales Price etc. has numeric values. But, it showing as object. These values has to converted into numeric values for calculation.

```python
columns = ["Units Sold","Manufacturing Price","Sale Price","Gross Sales","Discounts","Sales","COGS","Profit"]

for i in columns:
    data[i] = data[i].str.replace("$" , "" , regex = True)
```

```python
data.head()
```

### Now discount column has some "-" values. It has to change with zero for calculation.

```python
data["Discounts"] = data["Discounts"].replace("-","0",regex = True)
```

```python
data.dtypes
```

```python
columns = ["Units Sold","Manufacturing Price","Sale Price","Gross Sales","Discounts","Sales","COGS","Profit"]

for i in columns:
    data[i] = data[i].str.replace("," , "" , regex = True)
    
## Need to replace all the commas from numeric columns for calculation...
```

```python
data.head()
```

```python
data.dtypes
```

### Profit column has "  -    " values. It has to excluded from Profit column.

```python
## Data has soome string values in Profit column..
## Need to clean data in Profit column
(data["Profit"] == ' -   ').sum()
```

```python
data[data["Profit"] != " -   "]
```

```python
data = data[data["Profit"] != " -   "]
```

```python
data["Profit"] = data["Profit"].str.strip()
```

```python
data["Profit"].tail(50)
```

```python
data["Profit"] = data["Profit"].replace(r"[()]" , "" , regex = True)
```

```python
data.dtypes
```

### Now it can convert to numeric to float after cleaning values in Profit and other columns

```python
data["Profit"] = data["Profit"].astype("float")
```

```python
data["Units Sold"] = data["Units Sold"].astype("float")
data["Manufacturing Price"] = data["Manufacturing Price"].astype("float")
data["Sale Price"] = data["Sale Price"].astype("float")
data["Gross Sales"] = data["Gross Sales"].astype("float")
data["Sales"] = data["Sales"].astype("float")
data["Discounts"] = data["Discounts"].astype("float")
```

### Gross Margin on Products :

```python
py.figure(figsize=(10,6))

py.pie(product_group["Profit"] , labels = product_group["Product"] , autopct = "%1.1f%%")
py.title("Product-wise Profit" , fontsize = 20 , color = "red")
py.show()
```
